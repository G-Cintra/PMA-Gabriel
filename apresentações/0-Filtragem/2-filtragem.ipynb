```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# Configurações iniciais
np.random.seed(0)
anos = np.arange(2000, 2026, 1/12)
plt.rc('text', usetex=True)
# Série temporal com tendência, ciclos e ruído
tendencia = 100+0.02 * (anos - 2000)**2
ciclos = 3 * np.sin(4 * np.pi * (anos - 2000) / 10-np.pi/2)
sazonalidade_curto_prazo = 0.8 * np.sin(2 * np.pi * (anos - 2000) / 1)
ruido = 0.75 * np.random.normal(0, 1, len(anos))
serie_temporal = tendencia + ciclos + sazonalidade_curto_prazo + ruido

# Amostragem
anos_5anos = anos[::60]
serie_5anos = serie_temporal[::60]
anos_1ano = anos[::12]
serie_1ano = serie_temporal[::12]
anos_3meses = anos[::4]
serie_3meses = serie_temporal[::4]
# Centralizando o tempo em torno do ano inicial
t_centralizado = anos_5anos - anos_5anos[0]
t_centralizado1 = anos_1ano - anos_1ano[0]
t_centralizado2 = anos_3meses - anos_3meses[0]
# Transformar a série para log para ajustar uma tendência linear
log_serie_5anos = np.log(serie_5anos)
log_serie_1ano = np.log(serie_1ano)
log_serie_3meses = np.log(serie_3meses)
# Adiciona uma constante para o termo de intercepto
X = sm.add_constant(t_centralizado)
X1 = sm.add_constant(t_centralizado1)
X2 = sm.add_constant(t_centralizado2)
# Ajuste do modelo OLS na série log-transformada
model = sm.OLS(log_serie_5anos, X)
results = model.fit()
model2 = sm.OLS(log_serie_1ano, X1)
results2 = model2.fit()
model3 = sm.OLS(log_serie_3meses, X2)
results3 = model3.fit()
# Extraindo os parâmetros e convertendo-os
alpha_0, alpha_1 = results.params
y_0 = np.exp(alpha_0)  # Valor inicial da série exponencial
g_t = alpha_1          # Taxa de crescimento
alpha_01, alpha_11 = results2.params
y_01 = np.exp(alpha_01)  # Valor inicial da série exponencial
g_t1 = alpha_11          # Taxa de crescimento
alpha_02, alpha_12 = results3.params
y_02 = np.exp(alpha_02)  # Valor inicial da série exponencial
g_t2 = alpha_12          # Taxa de crescimento
print("Valor inicial (y_0):", y_0)
print("Taxa de crescimento (g_t):", g_t)
print("Valor inicial (y_01):", y_01)
print("Taxa de crescimento (g_t1):", g_t1)
print("Valor inicial (y_01):", y_02)
print("Taxa de crescimento (g_t1):", g_t2)

tendencia_ = alpha_0 + alpha_1*t_centralizado
tendencia_exponencial = y_0 * (1+g_t)**t_centralizado

tendencia_1 = alpha_01 + alpha_11*t_centralizado1
tendencia_exponencial1 = y_01 * (1+g_t1)**t_centralizado1

tendencia_2 = alpha_02 + alpha_12*t_centralizado2
tendencia_exponencial2 = y_02 * (1+g_t2)**t_centralizado2
# Set Seaborn style
sns.set(style="ticks")

```
